# **Hill Valley Dataset**

## **Objective**
Each record represents 100 points on a 2D graph. When plotted in order (from 1 to 100) as the Y coordinate, the points will create either a Hill or a Valley.

## **Data Source**
Data is taken from the YBI-Foundation Dataset repo.

```python
import pandas as pd
```

```python
hill = pd.read_csv('https://raw.githubusercontent.com/YBIFoundation/Dataset/main/Hill%20Valley%20Dataset.csv')
```

```python
hill.head()
```

```python
hill.describe()
```

```python
hill.info()
```

```python
hill.columns
```

```python
x = hill.drop(['Class'], axis = 1)
y = hill['Class']
```

```python
x.head()
```

```python
y.head()
```

```python
x.shape, y.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
x = ss.fit_transform(x)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 2529)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(max_iter = 5000)
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
y_pred_prob = model.predict_proba(x_test)
```

```python
y_pred_prob
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

```python
accuracy_score(y_test, y_pred)
```
